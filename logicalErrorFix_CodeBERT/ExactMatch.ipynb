```python
import pandas as pd
gold_path = "./model/cpp/checkpoint-best-bleu/test_0.gold"
output_path = "./model/cpp/checkpoint-best-bleu/test_0.output"

columns = ['index', 'code']

pd_gold = pd.read_csv(gold_path, sep = '\t', names=columns)
pd_output = pd.read_csv(output_path, sep = '\t', names=columns)
```

```python
print(pd_gold.head(10))
```

```python
print(pd_output.head(10))
```

```python
code_len = len(pd_gold)
exact_match_code = 0
exact_match_index = []

for index in range(code_len):
    if pd_gold.loc[index]["code"] == pd_output.loc[index]["code"]:
        exact_match_code += 1
        exact_match_index.append(index)

print(f"Exact Match Ratio { exact_match_code/code_len }")
print(exact_match_index)
```

```python
code_len = len(pd_gold)
error_line_detection = 0
error_line_detection_index = []

for index in range(code_len):
    if pd_gold.loc[index]["code"].split()[0] == pd_output.loc[index]["code"].split()[0]:
      error_line_detection += 1
      error_line_detection_index.append(index)

print(f"Error Line Detection Ratio { error_line_detection/code_len }")
print(error_line_detection_index)
```
